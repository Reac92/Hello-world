```python
def runningMean(x, N):
    y = np.zeros((len(x),))
    l=N
    l2=[]
    y[0]=np.sum(x[:N])
    for ctr in range(1,len(x)):
        l=l+N
        l2.append(x[l])
        if l<=len(x):
            y[ctr] = np.sum(x[ctr:N])+np.sum(l2)
    return y/N
N=runningMean(data[2][:1000],30)
print N
#plt.plot(range(0,len(N)),N)
#for ctr in N:
    
```
